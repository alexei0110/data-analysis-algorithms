```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calc_mae(y, y_pred):
    err = np.mean(np.abs(y - y_pred))
    return err

def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

###  Задание 1
Подберите скорость обучения (alpha) и количество итераций

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 55, 60, 35, 75, 80, 50, 60]
n = X.shape[0]

eta = 0.1 
n_iter = 1000

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}')
```

### Задание 3 
Вместо того чтобы задавать количество итераций, задайте условие остановки алгоритма, когда ошибка за итерацию начинает изменяться ниже определённого порога — упрощённый аналог параметра tol в линейной регрессии в sklearn.

```python
def mserror(X, w, y):
    y_pred = X.dot(w)
    return (np.sum((y_pred - y)**2)) / len(y)
X = np.array([[ 1,  1],              
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
Y = np.array([45, 55, 50, 55, 60, 35, 75, 80, 50, 60])
W = np.array([1, 0.5])
weight_dist = np.inf
min_weight_dist = 1e-8
eta = 0.01
errors = []
iter_num = 0

while weight_dist > min_weight_dist:
    y_pred = np.dot(X, W)
    dQ = 2 / Y.shape[0] * np.dot(X.T, y_pred - Y)
    new_w = W - eta * dQ
    weight_dist = np.linalg.norm(new_w - W, ord=2)
    error = mserror(X, new_w, Y)        
    errors.append(error)
    
    print(f'Iter {iter_num}: error - {error}, weights: {new_w}')
    
    iter_num += 1
    W = new_w   

print(f'В случае использования градиентного спуска функционал ошибки составляет {round(errors[-1], 4)}')
```
